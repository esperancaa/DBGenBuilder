```python
search_url = 'https://www.ncbi.nlm.nih.gov/nuccore?term=yeast'
r = requests.get(search_url)
soup = BeautifulSoup(r.content, 'html.parser')

# Find the links to the individual pages for each search result
results = soup.find_all('meta', {'name':"ncbi_uidlist"})
for result in results:
    # Get the link for each search result
    link = result['href']
    # Get the page for the individual search result
    individual_page = requests.get(link)
    soup = BeautifulSoup(individual_page.content, 'html.parser')
    # Do something with the content of the page
    print(soup)
```
